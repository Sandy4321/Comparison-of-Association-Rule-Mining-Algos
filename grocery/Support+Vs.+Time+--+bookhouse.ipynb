```python
import sys
from itertools import chain, combinations
from collections import defaultdict
from optparse import OptionParser
from fim import apriori
from fim import fpgrowth
from fim import eclat
import time
```

```python
file_iter = open('/home/vanessa/DWDM/bookhouse.csv', 'rU')
tracts = []
for line in file_iter:
    line = line.strip().rstrip(',')                         # Remove trailing comma
    record = line.split(',')
#    print record
    tracts.append(record)
#print("-------------------")
#print tracts
```

```python
xAxisap=[]
yAxisap=[]
for i in xrange(5, 50, 1):
    yAxisap.append(i)
    print  ('apriori(tracts, supp=',i,', zmin=2):')
    start = time.time()
    #print i
    for r in apriori(tracts, target = 's', supp = i, conf = 10, zmin=1): 
        print (r) 
    xAxisap.append(time.time()-start)
```

```python
import pyfpgrowth
xAxisfp=[]
yAxisfp=[]
for i in xrange(5, 50, 1):
    yAxisfp.append(i)
    print  ('apriori(tracts, supp=',i,', zmin=2):')
    start = time.time()
    #print i
    for r in fpgrowth(tracts, target = 'r', supp = i, conf = 10, zmin=2): 
        print (r) 
    #patterns = pyfpgrowth.find_frequent_patterns(tracts, i)
    #rules = pyfpgrowth.generate_association_rules(pyfpgrowth.find_frequent_patterns(tracts, i), 0.5)
    #print (rules)
    xAxisfp.append(time.time()-start)
```

```python
xAxisec=[]
yAxisec=[]
for i in xrange(5, 50, 1):
    yAxisec.append(i)
    print  ('apriori(tracts, supp=',i,', zmin=2):')
    start = time.time()
    #print i
    for r in eclat(tracts, target = 's', supp = i, conf = 10, zmin=1): 
        print (r) 
    xAxisec.append(time.time()-start)
```
