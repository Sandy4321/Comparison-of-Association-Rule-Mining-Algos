```python
import sys
from itertools import chain, combinations
from collections import defaultdict
from optparse import OptionParser
from fim import apriori
from fim import fpgrowth
from fim import eclat
import time
```

```python
file_iter = open('/home/vanessa/DWDM/bookhouse.csv', 'rU')
tracts = []
for line in file_iter:
    line = line.strip().rstrip(',')                         # Remove trailing comma
    record = line.split(',')
#    print record
    tracts.append(record)
```

```python
xAxisap=[]
yAxisap=[]
for i in xrange(5, 50, 7):
    yAxisap.append(i)
    print  ('apriori(tracts, supp=',i,', zmin=1):')
    start = time.time()
    #print i
    for r in apriori(tracts, target = 's', supp = -i, conf = 10, zmin=2): 
        print (r) 
    xAxisap.append(time.time()-start)
```
