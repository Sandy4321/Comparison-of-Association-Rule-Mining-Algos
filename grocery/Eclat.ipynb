```python
file_iter = open('/home/vanessa/DWDM/bookhouse.csv', 'rU')
tracts = []
for line in file_iter:
    line = line.strip().rstrip(',')                         # Remove trailing comma
    record = line.split(',')
#    print record
    tracts.append(record)
```

```python
xAxisec=[]
yAxisec=[]
for i in xrange(5, 50, 7):
    yAxisec.append(i)
    print  ('eclat(tracts, supp=',i,', zmin=1):')
    start = time.time()
    #print i
    for r in eclat(tracts, target = 's', supp = -i, conf = 10, zmin=2): 
        print (r) 
    xAxisec.append(time.time()-start)
```
